```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
## we will clean some of the data
dataframe = pd.read_csv('train_u6lujuX_CVtuZ9i.csv')
dataframe.drop(['Gender','Loan_ID'], axis=1, inplace=True)
dataframe.info()
```

```python
dataframe['CoapplicantIncome'] = dataframe['CoapplicantIncome'].fillna(dataframe['CoapplicantIncome'].median())
dataframe['LoanAmount'] = dataframe['LoanAmount'].fillna(dataframe['LoanAmount'].median())
dataframe.head(20)
```

```python
dataframe.info()
```

```python
dataframe.dropna(inplace=True)
dataframe.info()
```

Preparing Data

```python
from sklearn.preprocessing import OneHotEncoder
categorical_features = ['Married', 'Dependents', 'Education', 'Self_Employed','Property_Area']
x = dataframe.drop(['Loan_Status'], axis=1)
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output= False).set_output(transform= 'pandas')
ohe_x = ohe.fit_transform(x[categorical_features])
x = pd.concat([x, ohe_x], axis=1).drop(categorical_features,axis=1)
x.head(5)
```

```python
## maybe use OrdinalEncoder instead 
y = dataframe[['Loan_Status']]
y['Loan_Status'] = y['Loan_Status'].replace({'Y': 1, 'N': 0})
y.head()
```

```python
y.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=32)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
sel = SelectKBest(chi2, k=5)
x_new = sel.fit_transform(x, y)
x_new.shape
```

```python
selected_features = x.columns[sel.get_support()]
print("Selected Features:", list(selected_features))
```

```python
X_train = X_train[selected_features]
#X_train.shape
#how many credit history with the value 1 have a loan status of 0
```

```python
X_train
```

```python
X_test = X_test[selected_features]
X_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)  # Use same scaler for test set
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

K-nearest Neighbor




```python
from sklearn.neighbors import KNeighborsClassifier
dtc = KNeighborsClassifier(n_neighbors=15)
dtc.fit(X_train, y_train.values.ravel())
```

```python
training_score = dtc.score(X_train, y_train.values.ravel())
print("Training Score:", training_score)
test_score =  dtc.score(X_test, y_test.values.ravel())
print("Test Score:", test_score)
```

```python
y_pred = dtc.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
K = []
training = []
test = []
scores = {}
for k in range(1,20):
    dtc = KNeighborsClassifier(n_neighbors=k)
    dtc.fit(X_train, y_train.values.ravel())
    training_score = dtc.score(X_train, y_train.values.ravel())
    testing_score = dtc.score(X_test, y_test.values.ravel())
    K.append(k)
    training.append(training_score)
    test.append(testing_score)
    scores[k] = [training_score, testing_score]
```

```python
for keys, values in scores.items():
    print(keys, ':', values)
```

LOGISTIC REGRESSION





```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=32, class_weight = 'balanced', max_iter=1000)
clf.fit(X_train, y_train.values.ravel())
```

```python
clf.classes_
```

```python
from sklearn import metrics
y_pred_clf = clf.predict(X_test)
print (metrics.accuracy_score(y_test, y_pred_clf))
```

```python
cm = confusion_matrix(y_test, y_pred_clf)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_clf))
```

```python
proba = clf.predict_proba(X_test)
#we need to have a probability higher than threshold% to predict a loan status of 1 (need to reduce false positives)
threshold = 0.57
#15
# get predictions and convert to integers
custom_pred = (proba[:,1] > threshold).astype(int)
print(confusion_matrix(y_test, custom_pred))
```

```python
print("Custom Predictions:", custom_pred)
```

```python
print (classification_report(y_test, custom_pred))
```

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=5, shuffle=True, random_state=32)
model = LogisticRegression(random_state=0)
log_reg_scores = cross_val_score(model, X_train, y_train.values.ravel(), cv=kf, scoring='accuracy')
print(log_reg_scores)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
K = []
training = []
test = []
scores = {}
for k in range(1,10):
    dtc = DecisionTreeClassifier(random_state=32, max_depth= k)
    dtc.fit(X_train, y_train.values.ravel())
    training_score = dtc.score(X_train, y_train.values.ravel())
    testing_score = dtc.score(X_test, y_test.values.ravel())
    K.append(k)
    training.append(training_score)
    test.append(testing_score)
    scores[k] = [training_score, testing_score]
```

```python
for keys, values in scores.items():
    print(keys, ':', values)
```

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=5, shuffle=True, random_state=32)
tree = DecisionTreeClassifier(random_state=32, max_depth=3)
log_reg_scores = cross_val_score(tree, X_train, y_train.values.ravel(), cv=kf, scoring='accuracy')
print(log_reg_scores)
```

SVM

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear',C=.1,gamma=1, random_state=32)
svc.fit(X_train, y_train.values.ravel())
```

```python
for i in range(1,10):
    svc = SVC(kernel='linear',C=100,gamma=i, random_state=32)
    svc.fit(X_train, y_train.values.ravel())
    training_score = svc.score(X_train, y_train.values.ravel())
    testing_score = svc.score(X_test, y_test.values.ravel())
    K.append(i)
    training.append(training_score)
    test.append(testing_score)
    scores[i] = [training_score, testing_score]
    print(i, ':', [training_score, testing_score])
```

```python
score = svc.score(X_test, y_test.values.ravel())
print("Test Score:", score)
```
